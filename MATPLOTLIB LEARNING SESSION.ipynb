# 1.simple bar chart

```python
import matplotlib.pyplot as plt
plt.bar([1,2,3],[5,2,1],width=0.1,color="red")
plt.xlabel("This is x-axis label")
plt.ylabel("This is y-axis label")
plt.title("Simple Bar Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
student_name= [ "John","Sahil","Monica","Lara" ]
marks = [75,98,53,70]
plt.bar(student_name,marks)
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.title("Student Performance")
```

# 2.simple pie chart

```python
import matplotlib.pyplot as plt
import numpy as np
percentage = [60,30,10]
language = ["Python","Java","C++"]
plt.pie(percentage,labels=language)
plt.title("Simple Pie Chart")
plt.show()
```

# 3.simple scattering

```python
import matplotlib.pyplot as plt
import numpy as np
x= np.random.rand(100)
y=np.random.rand(100)
plt.title("Simple Scatter Plot")
plt.scatter(x,y,100,"green")
plt.show()
```

# 4.(ii) Area Chart Plotting (will use the function fill_between(x,y) to fill area b/w line(graph) & x-axis)

```python
import matplotlib.pyplot as plt #import the Python Matplotlib sub-module for graph plotting pyplot.
x = [1,2,3,4] # x axis
y = [1,2,3,4] # y axis
import matplotlib.pyplot as plt #import the Python Matplotlib sub-module for graph plotting pyplot.
x = [1,2,3,4] # x axis
y = [1,2,3,4] # y axis
plt.plot(x,y)# plotting the graph
plt.fill_between(x,y)
plt.show() #Displaying the figures
```

# 5.(i) multiple line plot

```python
import matplotlib.pyplot as plt
import numpy as np
x =np.arange(1,10) # values from 1 to 10
y1 = x+2
y2 = x*6
y3 = x/5
plt.plot(x,y1, x,y2, x,y3) # plot the figure
plt.show()
```

# 5.(ii) multiple line plot (colour specific)

```python
import matplotlib.pyplot as plt
import numpy as np
x =np.arange(1,10)# values from 1 to 10
y1 = x+2 
y2 = x*6
y3 = x/5
plt.plot(x,y1,color="red")
plt.plot(x,y2,color="pink")
plt.plot(x,y3,color="black")
plt.xlabel("years")
plt.ylabel("amount")
plt.title("financial conditions")
plt.grid("True")
plt.show()
```

# 6.Histogram Charts Plotting

```python
import matplotlib.pyplot as plt
import numpy as np
x= np.random.rand(100)
plt.title("Simple Histogram Chart")
plt.hist(x,10)
plt.show()
```

# 7.Any random function plot

```python
import matplotlib.pyplot as plt #import the Python Matplotlib sub-module for graph plotting pyplot
import numpy as np  #import the numpy module
x =np.arange(0.0,6.0,0.01)
y = [2**x for x in x]
plt.plot(x,y)# plotting the graph 
plt.show() #Displaying the figures
```
